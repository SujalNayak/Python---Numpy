```python
a = np.array([1,2,3,4,5,6])
print("Array a = ", a)
```

Reshape Of an Array

```python
a = np.arange(20)
a
```

```python
a.shape
```

```python
a.reshape(4,5) #2d array
```

```python
a.reshape(5, 2, 2)  #3d array
```

```python
a.reshape(-1,4)  #automatically calculates number of rows
a.reshape(2,-1)  #automatically calculates number of columns
```

Flatten Array - 
Converts all multi dimensional array to 1-d array

```python
a = np.arange(12).reshape(3,4)
a
```

```python
a.flatten()  #converts multi-dim array to 1d array
```

```python
a.flatten(order='F')  #column-wise flattening
```

```python
a.flatten(order='C')  #row-wise flattening
```

```python
a.flatten(order='K') #as per memory layout
```
